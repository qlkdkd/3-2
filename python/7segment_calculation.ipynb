# 2. +, -, x, / 네 가지 기호를 구현해보자. (기호당 2.5점, 총 10점)

```python
import math
```

```python
# + 그리기
def plus():
  forward(50)
  right(180)
  forward(25)
  right(90)
  forward(25)
  right(180)
  forward(50)
  left(90)
  penup()
  goto(400,250)

initializeTurtle()
plus()
```

```python
# - 그리기
def minus():
  right(90)
  forward(50)
  left(90)
  penup()
  goto(400,250)

initializeTurtle()
minus()
```

```python
# X(*) 그리기
def multiply():
  right(45)
  forward(50*math.sqrt(2))
  right(135)
  penup()
  forward(50)
  right(135)
  pendown()
  forward(50*math.sqrt(2))
  right(45)
  penup()
  goto(400,250)

initializeTurtle()
multiply()
```

```python
# / 그리기
def divide():
  right(45)
  forward(50*math.sqrt(2))
  left(45)
  penup()
  goto(400,250)

initializeTurtle()
divide()
```

```python
def equal():
  right(90)
  forward(50)
  right(90)
  penup()
  forward(25)
  right(90)
  pendown()
  forward(50)
  right(90)
  penup()
  goto(400,250)

initializeTurtle()
equal()
```

# 3. 입력된 숫자를 받아 사칙연산을 수행하는 코드를 작성하고 ""6x7=42"" 결과를 띄워보자. (총 40점)

# 단, 입력된 숫자는 한자리 자연수로 제한하고, 나눗셈 연산은 하지 않으며,

# 입력된 숫자와 연산기호, 등호, 연산결과 모두 Colab turtle을 활용해서 그려야 합니다.



```python

initializeTurtle()

# 연산 수행 함수
def draw_calculation(num1, operator, num2):
  # 첫 번째 숫자 그리기
  penup()
  goto(100, 250)
  pendown()
  draw_number(num1)

  # 연산자 그리기
  penup()
  goto(200, 250)
  pendown()
  draw_operator(operator)

  # 두 번째 숫자 그리기
  penup()
  goto(300, 250)
  pendown()
  draw_number(num2)

  # '=' 기호 그리기
  penup()
  goto(400, 250)
  pendown()
  equal()

  # 결과 숫자 계산 및 그리기
  result = calculate(num1, operator, num2)
  penup()
  goto(500, 250)
  pendown()
  draw_number(result)

# 숫자 그리기 함수 호출 (두 자리 숫자 처리)
def draw_number(num):
  if num < 10:
    # 한 자리 숫자는 그대로 그리기
    draw_single_digit(num)
  else:
    # 두 자리 숫자는 각 자릿수를 따로 그리기
    tens = num // 10
    units = num % 10
    penup()
    goto(600, 250)
    pendown()
    draw_single_digit(tens)

    # 일의 자리 숫자 위치로 이동
    penup()
    goto(700, 250)
    pendown()  # 간격 조정
    pendown()
    draw_single_digit(units)

# 한 자리 숫자 그리기 함수 호출
def draw_single_digit(num):
  if num == 0:
    zero()
  elif num == 1:
    one()
  elif num == 2:
    two()
  elif num == 3:
    three()
  elif num == 4:
    four()
  elif num == 5:
    five()
  elif num == 6:
    six()
  elif num == 7:
    seven()
  elif num == 8:
    eight()
  elif num == 9:
    nine()

# 연산자 그리기 함수 호출
def draw_operator(operator):
  if operator == '+':
    plus()
  elif operator == '-':
    minus()
  elif operator == '*':
    multiply()
  elif operator == '/':
    divide()

# 계산 수행
def calculate(num1, operator, num2):
  if operator == '+':
    return num1 + num2
  elif operator == '-':
    return num1 - num2
  elif operator == '*':
    return num1 * num2
  elif operator == '/':
    return num1 // num2  # 정수 나눗셈

# 예시: 6 + 7 그리기
draw_calculation(6, '*', 7)
```

# 4. 본인만의 특별한 연산을 추가해보자. (총 20점)

# 연산의 형태, 연산 기호의 모양 모두 자유롭게 설정하면 됩니다.

# 단, 두 개 이상의 입력을 받아 출력을 내는 연산으로 제한합니다.

```python
#평균을 구하는 연산자로서 네모 모양으로 설정함. 단, 정수연산만 가능하기에 결과는 정수로만 나옴
def avg():
  for i in range(4):
    forward(50)
    right(90)

initializeTurtle()
avg()
```

```python
# avg를 추가한 최종 코드


initializeTurtle()

# 연산 수행 함수
def draw_calculation(num1, operator, num2):
  # 첫 번째 숫자 그리기
  penup()
  goto(100, 250)
  pendown()
  draw_number(num1)

  # 연산자 그리기
  penup()
  goto(200, 250)
  pendown()
  draw_operator(operator)

  # 두 번째 숫자 그리기
  penup()
  goto(300, 250)
  pendown()
  draw_number(num2)

  # '=' 기호 그리기
  penup()
  goto(400, 250)
  pendown()
  equal()

  # 결과 숫자 계산 및 그리기
  result = calculate(num1, operator, num2)
  penup()
  goto(500, 250)
  pendown()
  draw_number(result)

# 숫자 그리기 함수 호출 (두 자리 숫자 처리)
def draw_number(num):
  if num < 10:
    # 한 자리 숫자는 그대로 그리기
    draw_single_digit(num)
  else:
    # 두 자리 숫자는 각 자릿수를 따로 그리기
    tens = num // 10
    units = num % 10
    penup()
    goto(600, 250)
    pendown()
    draw_single_digit(tens)

    # 일의 자리 숫자 위치로 이동
    penup()
    goto(700, 250)
    pendown()  # 간격 조정
    pendown()
    draw_single_digit(units)

# 한 자리 숫자 그리기 함수 호출
def draw_single_digit(num):
  if num == 0:
    zero()
  elif num == 1:
    one()
  elif num == 2:
    two()
  elif num == 3:
    three()
  elif num == 4:
    four()
  elif num == 5:
    five()
  elif num == 6:
    six()
  elif num == 7:
    seven()
  elif num == 8:
    eight()
  elif num == 9:
    nine()

# 연산자 그리기 함수 호출
def draw_operator(operator):
  if operator == '+':
    plus()
  elif operator == '-':
    minus()
  elif operator == '*':
    multiply()
  elif operator == '/':
    divide()
  elif operator == 'avg':
    avg()

# 계산 수행
def calculate(num1, operator, num2):
  if operator == '+':
    return num1 + num2
  elif operator == '-':
    return num1 - num2
  elif operator == '*':
    return num1 * num2
  elif operator == '/':
    return num1 // num2  # 정수 나눗셈
  elif operator == 'avg':
    return (num1+num2)//2

# 예시: 6 + 7 그리기
draw_calculation(6, 'avg', 7)
```
