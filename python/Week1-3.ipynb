## *Q. 사용자가 시간을 초로 입력하면 입력된 초가 몇 시간 몇 분 몇 초에 해당하는지 출력하는 코드를 몫과 나머지 연산을 이용해 구현해보자*

````python
# prompt: 사용자가 시간을 초로 입력하면 입력된 초가 몇 시간 몇 분 몇 초에 해당하는지 출력하는 코드를 몫과 나머지 연산을 이용해 구현해

seconds = int(input("시간을 초로 입력하세요: "))

hours = seconds // 3600
seconds = seconds%3600

minutes = seconds // 60
seconds %= 60

print(f"{hours}시간 {minutes}분 {seconds}초")
````

## *Q. 원금이 a, 이자율이 r인 경우, n 년후의 원리금 합계는 $a(1+r)^n$ 이 된다. 파이썬 코드를 통해 원금과 이자율 및 저금 기간을 입력 했을 때 원리금 합계금액을 print하는 코드를 짜보자*

````python
# prompt: 원금이 a, 이자율이 r인 경우, n 년후의 원리금 합계는  a(1+r)n  이 된다. 파이썬 코드를 통해 원금과 이자율 및 저금 기간을 입력 했을 때 원리금 합계금액을 print해

a = int(input("원금을 입력하세요: "))
r = float(input("이자율을 입력하세요: "))
n = int(input("저금 기간(년)을 입력하세요: "))

total_amount = a * (1 + r) ** n

print(f"원리금 합계: {total_amount}")
````

## *Q. 근의 공식을 통해 방정식 $x^2-4x+1 = 0$ 의 근을 구해보자*
### $x=\frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$
### 같은 공식을 활용해 $x^2-x+4 = 0$ 의 근을 구해보자
### 나아가 ax^2 + bx + c = 0 이라는 방정식에 대해 a, b, c 값이 입력됐을 때 근을 print 하는 코드를 짜보자

### 이때 a, b, c input을 a, b, c = map(int, input("세 정수를 입력하세요").split()) 을 통해서 받도록 한다.

````python
a, b, c = map(int, input("세 정수를 입력하시오: ").split())
# ax^2 + bx + c = 0

# 두 근
x1 = (-b + (b**2 - 4*a*c)**(1/2)) / (2*a)
x2 = (-b - (b**2 - 4*a*c)**(1/2)) / (2*a)
print(x1, x2)
````

## *Q. 삼각형의 세변의 길이를 입력받아 삼각형의 넓이를 출력하는 코드를 짜보자*

````python
a, b, c = map(int, input("세 정수를 입력하시오: ").split())
# 세변의 길이를 입력받아 삼각형의 넓이를 출력하는 코드를 짜보자
````

````python
# prompt: 삼각형의 세 변의 길이를 가지고 넓이를 구해줘

import math

a, b, c = map(int, input("세 변의 길이를 입력하세요: ").split())

# s: 반둘레
s = (a + b + c) / 2

# 헤론의 공식을 이용하여 넓이 계산
area = (s * (s - a) * (s - b) * (s - c))**0.5

print("삼각형의 넓이:", area)
````

## *Q. input 함수를 이용해서 시작위치, 이동방향, 이동거리 를 입력받아 해당 동작을 수행하는 코드를 작성하세요*
