# 4. 조건을 따져 실행해보자
## 4.1. 프로그램의 기본 제어 구조는 세 가지가 있다.
* 프로그램의 세 가지 기본적인 제어구조
    * 순차구조: 여러 명령이 순차적으로 실행되는 구조
        * 자동차가 직진하는 도로로 비유
    * 선택구조: 여러 개 중 하나의 명령문을 선택하여 실행하는 구조
        * 자동차가 여러 가지의 길 중에서 하나를 선택하여 주행하는 도로로 비유
    * 반복 구조: 동일한 명령이 반복되면서 실행되는 구조
        * 자동차가 회전하면서 주행하는 회전 교차로로 비유

* 어떤 프로그램이라도 3가지의 기본 블록만 있으면 만들 수 있음

### 선택 구조가 필요한 이유
* 만약 선택 구조가 없다면 프로그램은 항상 동일한 동작만을 되풀이할 것임
* 프로그램이 항상 동일한 동작만 한다면 언제나 정해진 결론에만 도달할 것임

## 4.2. 조건이 맞을 때만 실행되는 if문
### 관계 연산자
* 관계 연산자의 결과는 True나 False로 나타낸다. -> 조건 연산자는 피연산자들의 크기를 비교하여 True 또는 False를 돌려줌
* 조건식에는 논리 연산자를 사용할 수도 있다. 논리 연산자는 여러 개의 부울형 데이터를 논리적으로 연산하여 최종적으로 참 또는 거짓의 값을 갖는 수식을 만들 수 있다.
* 이 논리 연산을 이용하여 하나의 관계 연산자가 사용된 단순한 조건식이 아니라 여러 개의 관계 연산자가 사용된 복합 조건식을 만들 수 있다.

* if를 이용한 간단한 선택 구조
```python
if 조건식:
    실행할 내용
```

* 다음과 같이 변수 x에 100을 넣어준 후 아래와 같이 if문을 작성해 보자. 이 경우 조건식에 해당하는 x>1이 True를 반환하므로 실행할 내용이 선택되어 실행된다.
* 따라서 화면에는 'x는 1보다 큽니다.'가 출력되는 것을 볼 수 있다. 
이 때 : 아래의 줄은 **반드시 들여쓰기**를 해야만 한다.

````python
x=100
if x>1:
    print("x는 1보다 큽니다.")
````

### 들여쓰기가 아주 중요한 파이썬: 블록은 들여쓰기로 완성
* 성적이 60점 이상이면 합격과 동시에 장학금도 받을 수 있다고 출력하려면 어떻게 해야 할까? 이런 경우에는 다음과 같이 **들여쓰기를 이용하여서** 문장들을 묶을 수 있다.

````python
score=int(input("점수 입력: "))
if score>=60:
    print('합격입니다.')
    print('장학금도 수령하세요')
````

* 코드 블록은 *특정한 동작을 위해서 코드가 모여있는 상태*를 뜻하는데 파이썬은 들여쓰기를 기준으로 코드 블록을 구성한다. 하나의 블록에 속하는 문장들은 특정 조건이 참일때 모두 같이 실행된다.
* 블록에 있는 문장들은 그 위에 있는 문장들과 비교할 때, 앞에 4칸의 들여쓰기 공백을 두고 있다.

* 파이썬에서 문장 앞에 동일한 수의 공백이 있으면 이들 문장들은 하나의 블록에 속하게 된다. 만약 동일한 블록에 속해야 하는데 실수로 공백을 더 많이 추가하였다면 오류가 발생한다.

````python
score=int(input("점수 입력: "))
if score>=60:
 print('합격입니다.')
  print('장학금도 수령하세요')
````

## 4-3. 배타적 조건에 따라 실행하는 if-else문
* 파이썬에는 선택 구조를 위하여 if-else문을 제공한다.

````python
score=int(input("성적 입력: "))
if score>=60:
    print("합격")
else:
    print("불합격")
````

* if-else문장은 "만약 조건이 참이면 이것을 실행하고, 조건이 참이 아니라면 저것을 실행해!"라고 말하는 것과 같다. if-else문에서는 조건을 수식으로 표현하며, 그 수식을 바로 '조건식'이라고 한다.
* 조건식 뒤에는 반드시 콜론(:)이 와야 한다. 콜론은 파이썬 인터프리터에게 "아직 전체 문장을 끝나지 않았으니 잠시 해석을 미뤄달라. 이 다음에는 이 문장에 딸린 콛 블록이 나타날 것이다."라고 요청하는 기호다.
* if-else문은 주어진 조건식을 계산하여 조건식이 참(True)으로 계산되면 if 아래에 있는 문장을 실행한다. 만약 거짓(False)이면 else아래에 있는 문장을 실행한다.

````python
score=int(input("성적 입력: "))
if score>=60:
    print("합격")
else:
    print("불합격")
````

* 또다른 예시로 24시 체계의 시간정보를 입력받아 12보다 작으면 '오전입니다.', 12 이상이면 '오후입니다.'를 출력하는 시계가 있다고 하자. 0에서 23까지의 시간이 입력될 경우 오전이면서 동시에 오후가 될 수 없으므로 이 조건은 **배타적 조건**이 되어 다음과 같은 if-else문으로 표현 가능하다.

````python
hour=10

if hour<12:
    print("오전입니다.")
else:
    print("오후입니다.")
````

## Lab4.5. 랜덤 함수로 동전 던지기 게임을 만들자
이 예제에서는 동전 던지기 게임을 작성해보자. 동전을 던지는 것은 난수를 생성하면 된다. 파이썬에서는 import random한 후에 random.randrange(2)와 같이 하면 0이나 1을 랜덤하게 생성할 수 있다.

````python
import random

print("동전 던지기 게임을 시작합니다.")
coin=random.randrange(2)
if coin==0:
    print('앞면입니다.')
else:
    print("뒷면입니다.")
print("게임이 종료되었습니다.")
````

#### 도전문제 4.9
주사위 던지기 게임으로 변환해보자. random.randint(1, 6)을 사용하면 1에서 6까지의 정수를 랜덤하게 생성할 수 있다. 입력으로 '홍길동', '홍길순'과 같은 사용자의 이름을 받아서 홍길동의 주사위 번호, 홍길순의 주사위 번호를 출력한 후 이들을 비교하여 다음과 같이 비교하여 다음과 같이 출력하여라. 결과는 '홍길동이 이겼습니다.', '홍길순이 이겼습니다.', '비겼습니다.' 가운데 하나가 나타나도록 하여라.

````python
import random

def print_user(p1, p2, p1_die, p2_die):
    print(f"{p1}의 주사위 번호는 {p1_die}")
    print(f"{p2}의 주사위 번호는 {p2_die}")

p1=input("Player1의 이름: ")
p2=input("Player2의 이름: ")
print('...주사위를 굴립니다...')
p1_die=random.randint(1, 6)
p2_die=random.randint(1, 6)
if (p1_die>p2_die):
    print_user(p1, p2, p1_die, p2_die)
    print(f'{p1}이 이겼습니다.')
elif (p1_die<p2_die):
    print_user(p1, p2, p1_die, p2_die)
    print(f"{p2}이 이겼습니다.")
else:
    print_user(p1, p2, p1_die, p2_die)
    print("비겼습니다.")
````

## 4.6. 조건이 거짓일 때 연속하여 다른 조건을 검사
* if-else문에서 조건이 거짓을 때 다른 조건을 검사할 수 있다. 이때는 elif키워드를 사용하면 된다.


````python
num=int(input("정수를 입력하시요: "))

if num>0:
    print("양수입니다.")
elif num==0:
    print("0입니다.")
else:
    print("음수입니다.")
````

### if-else문 안에 if-else문 넣기
* 필요에 따라 if문 안에 또 다른 if문에 들어갈 수도 있는데, 이것을 내부 if블록, 또는 중첩 if블록이라고 한다. 
* 특정한 상황에서 True라는 조건이 나왔을 경우, 이 상황에 대해서 또 다른 세부적인 조건을 검사하는 상황에서 발생할 수 있다.


````python
num=int(input("정수 입력: "))

if num>=0:
    if num==0:
        print("0입니다.")
    else:
        print("양수입니다.")
else:
    print("음수입니다.")
````
