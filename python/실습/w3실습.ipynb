## 몫과 나머지의 계산

```python
p=int(input("분자를 입력: "))
q=int(input("분모 입력: "))

print(f"나눗셈의 몫: {p//q}")
print(f"나눗셈의 나머지: {p%q}")
```

```python
a=3
b=4
print(f'곱하기: {a*b}')
print(f"거듭제곱: {a**b}")
```

```python
#q1
sec=int(input("초 입력: "))
min=sec//60
hour=min//60
sec%=60

print(f"{hour}:{min}:{sec}")
```

```python
#q2
a, r, n=map(int, input("원금, 이자율, 연도 입력").split())
print(f"원리금: {a*(1+r)**n}")
```

```python
#q3
a, b, c=map(int, input().split())
x1=(-b+(b**2-4*a*c)**(1/2))/(2*a)
x2=(-b-(b**2-4*a*c)**(1/2))/(2*a)

print(x1, x2)
```

```python
#q4
a, b, c=map(int, input().split())
s=(a+b+c)/2
r=(s*(s-a)*(s-b)*(s-c))**(1/2)
print(r)
```
