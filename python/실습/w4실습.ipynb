### 조건문 if의 활용
* if-else
* if-elif-else

```python
#A학점의 기준, 조건문
score=int(input())
if score>=90:
  print("A")
```

```python
#들여스기의 정도는 상관 없음(colab기준 2칸)
score=int(input())
if score>=90:
  print("A")
```

```python
#나머지 학생들에 대한 학점
score=int(input())
if score>=90:
  print('A')
else:
  print('B')
```

```python
#elif를 통해 조건 추가
score=int(input())
if score>=90:
  print("A")
elif score>=80:
  print("B")
else:
  print("C")
```

```python
#elif를 쓰지 않고 표현
score=int(input())
if score>=90:
  print("A")
else:
  if score>=80:
    print("B")
  else:
    print("C")
```

```python
#if만을 사용
score=int(input())
if score>=90:
  print('A')
if score>=80 and score<90:
  print('B')
if score<80:
  print('c')
```

## 덧셈 뺄셈 곱셈 나눗셈 수행하는 프로그램

```python
in1, in2=map(int, input("연산할 숫자 두개 입력: ").split())
oper=input("연산자 입력:")

if oper=="+":
  print(in1+in2)
elif oper=='-':
  print(in1-in2)
elif oper=='*':
  print(in1*in2)
elif oper=='/':
  print(in1/in2)
```

## BMI 계산기 만들기

```python
height, weight=map(float, input("키(cm)와 몸무게(kg) 입력: ").split())
BMI=weight/(height/100)**2

if BMI<18.5:
  print("저체중")
elif BMI<23:print("정상")
elif BMI>=23 and BMI<25: print("과체중")
else: print("비만")
```

## Ramdom모듈을 활용해서 복권 문제를 만들어보자

```python
import random

lottery_numbers1=random.randint(0, 9)
lottery_numbers2=random.randint(0, 9)
print(lottery_numbers1, lottery_numbers2)

num1, num2=map(int, input("숫자 2개 입력: ").split())
if num1==lottery_numbers1 and num2==lottery_numbers2:
  print("두 숫자 맞춤")
elif num1==lottery_numbers1 and num2==lottery_numbers1:
  print("한 숫자 맞춤")
elif num1==lottery_numbers2 and num2==lottery_numbers2:
  print("한 숫자 맞춤")
else:
  print("다틀림")
```

## Q. 삼각형 사각형 중 하나를 입력받아 도형을 그리는 코드를 작성해보자.

```python
shape_input=input()
s=int(input())

initializeTurtle()
if shape_input=="사각형":
  for i in range(4):
    fd(s)
    right(90)
elif shape_input=="삼각형":
  for i in range(3):
    fd(s)
    right(120)
```

## Q. 위의 문제에 이어서 거북이를 임의의 위치로 이동시켜서 해당 도형(사각형부터) 내부에 있는지 확인하는 코드를 만들어보자

```python
import random
initializeTurtle()
penup()
goto(0, 0)
pendown()

for i in range(4):
  right(90)
  fd(200)


randx=random.randint(0, 200)
randy=random.randint(0, 200)
goto(randx, randy)

if (randx>0 and randx<200) and (randy>0 and randy<200):
  print('도형 안')
else: print('도형 밖')
```
